10. What is the T-distribution, and when should it be used instead of the normal distribution?


Ans:The t-distribution, also known as Student's t-distribution, is a probability distribution used in statistics, particularly when dealing with small sample sizes or when the population standard deviation is unknown. It is preferred over the normal distribution in these situations because it accounts for the increased uncertainty associated with estimating population parameters from limited data.

Here's a more detailed explanation:

What is the t-distribution:

 - The t-distribution is similar to the normal distribution (also called the z-distribution) in that it is bell-shaped and symmetric around a mean of zero.

 - However, the t-distribution has "heavier tails," meaning it has a greater probability of extreme values than the normal distribution.

 - The shape of the t-distribution is determined by its degrees of freedom, which are related to the sample size.

 - As the sample size (and thus the degrees of freedom) increases, the t-distribution gradually approaches the normal distribution.

When to use the t-distribution instead of the normal distribution:

 - Small sample sizes:
When working with a small sample (generally considered to be less than 30), the t-distribution is more appropriate for estimating population parameters. This is because the sample standard deviation is a less reliable estimate of the population standard deviation when the sample size is small.

 - Unknown population standard deviation:
If the population standard deviation is unknown, the t-distribution is used to account for the uncertainty in estimating it from the sample standard deviation.

 - Confidence intervals and hypothesis testing:
The t-distribution is used in constructing confidence intervals for population means and in performing t-tests, which are used to compare the means of two samples.

 - The t-distribution provides a more accurate representation of the data when dealing with uncertainty related to small sample sizes or unknown population standard deviations. The normal distribution is more suitable when the sample size is large and the population standard deviation is known.

It allows statisticians to combine prior knowledge with observed data to refine their understanding of probabilities, making it a crucial tool for inference and decision-making.

Bayes' Theorem in Statistics:

Bayes' theorem is expressed mathematically as:
P(A|B) = [P(B|A) * P(A)] / P(B)

Where:

 - P(A|B)
is the posterior probability: The probability of event A happening given that event B has already occurred.

 - P(B|A)
is the likelihood: The probability of event B happening given that event A has already occurred.

 - P(A)
is the prior probability: The initial probability of event A before considering any new evidence.

 - P(B)
is the marginal probability: The probability of event B occurring, regardless of event A.

Significance of Bayes' Theorem:

1. Updating Probabilities:
Bayes' theorem is significant because it allows us to update our beliefs or probabilities as new evidence becomes available. We start with a prior probability, and as we observe new data, we can calculate a posterior probability that reflects the updated likelihood of an event.

2. Inversion of Conditional Probabilities:
It allows us to calculate the probability of a cause (event A) given its effect (event B), even if it's easier to directly observe the probability of the effect given the cause (B|A).

3. Bayesian Inference:
Bayes' theorem is foundational to Bayesian inference, a statistical approach that uses Bayes' theorem to update probabilities of hypotheses or models based on observed data.

4. Applications in Various Fields:

Bayes' theorem has diverse applications, including:

 - Medical Diagnosis: Determining the probability of a disease given symptoms or test results.

 - Spam Filtering: Classifying emails as spam based on the presence of certain words.

 - Machine Learning: Used in algorithms like Naive Bayes for classification and spam filtering.

 - Financial Modeling: Assessing investment risks and returns.

 - Risk Assessment: Evaluating the probability of various risks in different scenarios.

 - Anomaly Detection: Identifying unusual patterns or outliers in datasets.

 - Personalization: Recommending products or content based on user preferences.

 - Example:
Imagine a medical test for a rare disease.

 - P(Disease):
The prior probability of having the disease (based on its prevalence in the population).

 - P(Positive Test | Disease):
The probability of testing positive given that you have the disease (test accuracy).

 - P(Positive Test):
The probability of a positive test result (calculated using both true positives and false positives).

Bayes' theorem allows us to calculate P(Disease | Positive Test), which is the probability of actually having the disease given a positive test result. This is crucial because a positive test result doesn't necessarily mean a high probability of having the disease, especially for rare diseases.

It assesses whether there's a significant difference between observed frequencies (what you actually see in your data) and expected frequencies (what you would anticipate based on a specific distribution). Essentially, it checks if the observed data "fits" the expected distribution.

Here's how it's applied:

1. Define the Hypothesis:

 - Null Hypothesis (H0): The observed distribution matches the expected distribution.

 - Alternative Hypothesis (Ha): The observed distribution does not match the expected distribution.

2. Collect and Organize Data: Gather your observed data, grouping it into categories or intervals.

3. Calculate Expected Frequencies: For each category, determine the expected frequency based on the hypothesized distribution.

4. Calculate the Chi-Square Statistic: Use the formula: Ï‡Â² = Î£ [(Observed - Expected)Â² / Expected]. This formula calculates the sum of squared differences between observed and expected frequencies, divided by the expected frequencies.

5.  Determine Degrees of Freedom: Degrees of freedom (df) are calculated as (number of categories - 1).

6. Find the P-value: Using the calculated chi-square statistic and degrees of freedom, find the corresponding p-value from a chi-square distribution table or statistical software.

7. Make a Decision:

 - If the p-value is less than your chosen significance level (alpha, often 0.05), reject the null hypothesis. This suggests a statistically significant difference between the observed and expected distributions.

 - If the p-value is greater than or equal to the significance level, fail to reject the null hypothesis. This indicates that the observed data is consistent with the expected distribution.
  
In simpler terms, the chi-square goodness-of-fit test helps you decide if the differences you see in your data are likely due to random chance or if they suggest a real discrepancy between what you observed and what you expected.

ANOVA is widely used in various fields, including biomedical research, to analyze data from experiments with multiple groups.

Assumptions of ANOVA:

To ensure the validity of ANOVA results, several key assumptions need to be met:

1. Normality:
The data within each group should be approximately normally distributed. This means that the data should follow a bell-shaped curve when plotted.

2. Homogeneity of Variance:
The variances (or standard deviations) of the groups being compared should be roughly equal. This is also known as homoscedasticity.

3. Independence:
The observations within each group and between groups should be independent of each other. This means that the data collected from one group should not influence the data collected from another group.

Violating these assumptions can lead to inaccurate results, so it's important to check them before conducting an ANOVA test.


What Happens If Assumptions Fail:

 - Independence violation: Seriousâ€”requires different methods (e.g., repeated-measures ANOVA).

 - Normality/variance issues: ANOVA is robust if sample sizes are large and similar. Otherwise:

 - Use Welchâ€™s ANOVA (handles unequal variances),

 - Or apply data transformations (e.g., log transform).


 Typical Workflow:

1. Check assumptions (normality via Q-Q plots or tests; homogeneity via Leveneâ€™s).

2. Run ANOVA, obtaining F-statistic and p-value.

3. If p â‰¤ Î± (e.g., 0.05), reject null hypothesis (at least one mean differs).

```python
# 6.  Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?
import numpy as np
import matplotlib.pyplot as plt

def plot_discrete_uniform_cdf(lower_bound, upper_bound):

    if not isinstance(lower_bound, int) or not isinstance(upper_bound, int):
        raise TypeError("Bounds must be integers.")
    if lower_bound >= upper_bound:
        raise ValueError("Lower bound must be less than the upper bound.")

    # Generate the range of possible values for the discrete uniform distribution
    x_values = np.arange(lower_bound, upper_bound + 1)

    # Calculate the probability of each value (1/N, where N is the number of values)
    num_values = len(x_values)
    probability_per_value = 1 / num_values

    # Calculate the cumulative probabilities
    cdf_values = np.cumsum(np.full(num_values, probability_per_value))

    plot_x = np.concatenate(([lower_bound - 1], x_values, [upper_bound + 1]))
    plot_y = np.concatenate(([0], cdf_values, [1]))

    # Plot the CDF
    plt.figure(figsize=(8, 6))
    plt.step(plot_x, plot_y, where='post', label='CDF', linestyle='-', marker='o', markersize=4)
    plt.title(f'Cumulative Distribution Function of Discrete Uniform Distribution ({lower_bound} to {upper_bound})')
    plt.xlabel('Value (x)')
    plt.ylabel('F(x) = P(X <= x)')
    plt.grid(True)
    plt.yticks(np.arange(0, 1.1, 0.1)) # Set y-axis ticks for better readability
    plt.legend()
    plt.show()

# Example usage:
plot_discrete_uniform_cdf(1, 6) # Represents a fair six-sided die
# plot_discrete_uniform_cdf(5, 10)
```

```python
# 10  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.?
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def simulate_clt(pop_dist, sample_size, num_samples):
    """
    Simulate sampling from a population distribution `pop_dist`
    `num_samples` times, each with `sample_size` observations.
    Returns: array of sample means.
    """
    samples = pop_dist(size=(num_samples, sample_size))
    means = samples.mean(axis=1)
    return means

def plot_sampling_means(means, sample_size):
    plt.figure(figsize=(8, 5))
    sns.histplot(means, kde=True, bins=30, color='blue', edgecolor='black')
    plt.title(f'CLT: Sample Means (n={sample_size})')
    plt.xlabel('Sample mean')
    plt.ylabel('Density')

    # Overlay theoretical normal curve
    mu, sigma = means.mean(), means.std(ddof=1)
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 200)
    plt.plot(x, 1/(sigma * np.sqrt(2*np.pi)) *
             np.exp(-0.5*((x - mu)/sigma)**2),
             'r--', label='Normal approximation')
    plt.legend()
    plt.show()

def main():
    np.random.seed(42)
    pop_dist = lambda size: np.random.exponential(scale=2, size=size)

    sample_sizes = [1, 5, 30, 100]
    num_samples = 5000

    for n in sample_sizes:
        means = simulate_clt(pop_dist, sample_size=n, num_samples=num_samples)
        print(f"n = {n:3d} | sample means Âµ â‰ˆ {means.mean():.2f}, Ïƒ â‰ˆ {means.std(ddof=1):.2f}")
        plot_sampling_means(means, n)

if __name__ == "__main__":
    main()
```

```python
# 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_score(x, data):
    """
    Compute the Z-score of x relative to the dataset `data`.
    Z = (x - Î¼) / Ïƒ
    """
    mu = np.mean(data)
    sigma = np.std(data, ddof=0)  # population std dev
    z = (x - mu) / sigma
    return z, mu, sigma

def plot_z_on_standard_normal(z):
    """
    Plot the standard normal PDF and shade areas for P(Z <= z) and P(Z >= z).
    """
    x = np.linspace(-4, 4, 400)
    pdf = norm.pdf(x)
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, 'black', label='Standard Normal PDF')

    # Shade P(Z <= z)
    xs1 = np.linspace(-4, z, 200)
    plt.fill_between(xs1, norm.pdf(xs1), color='skyblue', alpha=0.5, label=f'P(Z â‰¤ {z:.2f})')

    # Shade P(Z >= z)
    xs2 = np.linspace(z, 4, 200)
    plt.fill_between(xs2, norm.pdf(xs2), color='salmon', alpha=0.5, label=f'P(Z â‰¥ {z:.2f})')

    plt.title('Standard Normal Distribution with Z-score Highlighted')
    plt.xlabel('Z')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

def analyze_point(x, data):
    z, mu, sigma = z_score(x, data)
    p_lower = norm.cdf(z)
    p_upper = 1 - p_lower

    print(f"Data point x = {x}")
    print(f"Dataset mean Î¼ = {mu:.3f}, Ïƒ = {sigma:.3f}")
    print(f"Z-score = {z:.3f} â†’ P(Z â‰¤ z) = {p_lower:.3f}, P(Z â‰¥ z) = {p_upper:.3f}")

    plot_z_on_standard_normal(z)

if __name__ == "__main__":
    # Example dataset
    data = np.random.normal(loc=50, scale=5, size=1000)
    x = 60  # data point to evaluate
    analyze_point(x, data)
```

```python
# 15.  Implement hypothesis testing using Z-statistics for a sample dataset?
import numpy as np
from statsmodels.stats.weightstats import ztest as ztest_function

def perform_one_sample_z_test(sample_data, population_mean_hypothesized, alpha=0.05, alternative='two-sided'):

    # Convert sample data to a NumPy array for calculations
    sample_data = np.array(sample_data)

    # Perform the Z-test
    # ztest_function returns (z_statistic, p_value)
    z_statistic, p_value = ztest_function(x1=sample_data, value=population_mean_hypothesized, alternative=alternative)

    # Determine the conclusion based on the p-value and significance level
    if p_value < alpha:
        conclusion = "Reject the null hypothesis."
    else:
        conclusion = "Fail to reject the null hypothesis."

    return z_statistic, p_value, conclusion

# Example Usage:
if __name__ == "__main__":
    # Sample data (e.g., test scores of a class)
    sample_scores = [85, 92, 78, 88, 95, 80, 82, 90, 75, 89]

    # Hypothesized population mean (e.g., average test score of all students historically)
    hypothesized_mean = 80

    # Significance level
    significance_level = 0.05

    # Perform the Z-test
    z_stat, p_val, test_conclusion = perform_one_sample_z_test(
        sample_scores, hypothesized_mean, significance_level
    )

    # Print the results
    print(f"Sample Data: {sample_scores}")
    print(f"Hypothesized Population Mean: {hypothesized_mean}")
    print(f"Significance Level (alpha): {significance_level}")
    print(f"\nZ-statistic: {z_stat:.4f}")
    print(f"P-value: {p_val:.4f}")
    print(f"Conclusion: {test_conclusion}")

    # Example with a different alternative hypothesis (e.g., 'larger')
    print("\n--- Testing with 'larger' alternative hypothesis ---")
    z_stat_larger, p_val_larger, test_conclusion_larger = perform_one_sample_z_test(
        sample_scores, hypothesized_mean, significance_level, alternative='larger'
    )
    print(f"Z-statistic (larger): {z_stat_larger:.4f}")
    print(f"P-value (larger): {p_val_larger:.4f}")
    print(f"Conclusion (larger): {test_conclusion_larger}")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result?

import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence=0.95):

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    n = len(data)

    # Calculate the t-value for the given confidence level and degrees of freedom
    # Degrees of freedom (df) = n - 1
    t_value = stats.t.ppf((1 + confidence) / 2, df=n - 1)

    # Calculate the margin of error
    margin_of_error = t_value * (sample_std / np.sqrt(n))

    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return sample_mean, lower_bound, upper_bound

# Example Usage:
data = [12, 15, 14, 10, 13, 17, 14, 15, 16, 14]
sample_mean, lower_bound, upper_bound = calculate_confidence_interval(data)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Interpretation of the result
print("\nInterpretation:")
print(f"We are 95% confident that the true population mean lies within the interval of ({lower_bound:.2f}, {upper_bound:.2f}).")
print("This means that if we were to repeat this sampling process many times,")
print("approximately 95% of the resulting confidence intervals would contain the true population mean.")
```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean?
import numpy as np
import scipy.stats as st

def confidence_interval(data, confidence=0.95):

    a = 1.0 * np.array(data)
    n = len(a)
    if n <= 1:
        return np.mean(a), np.mean(a), np.mean(a)
    mean = np.mean(a)
    sem = st.sem(a)  # Standard error of the mean
    h = sem * st.t.ppf((1 + confidence) / 2., n-1)
    return mean, mean-h, mean+h

# Generate sample data from a normal distribution
np.random.seed(0)  # for reproducibility
sample_size = 50
mean_val = 100
std_dev = 15
data = np.random.normal(mean_val, std_dev, sample_size)

# Calculate the confidence interval
mean, lower, upper = confidence_interval(data)

# Print the results
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")


# Interpretation:
# With 95% confidence, the true population mean lies between {lower:.2f} and {upper:.2f}.  This means that if we were to repeat this sampling process many times, 95% of the calculated confidence intervals would contain the true population mean.
```

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?

import numpy as np
from statsmodels.stats.weightstats import ztest

def z_test_sample(data, popmean, popstd, alpha=0.05, alternative='two-sided'):

    # Statsmodels ztest assumes unknown sigma; to simulate known sigma, scale data
    n = len(data)
    scaled_data = (data - popmean) / popstd * np.sqrt(n)

    z_stat, p_value = ztest(scaled_data, value=0, alternative=alternative)
    return z_stat, p_value

if __name__ == "__main__":
    np.random.seed(42)
    # Generate sample: n=50, true mean=100, pop std known=15
    popmean = 100
    popstd = 15
    n = 50
    data = np.random.normal(loc=102, scale=popstd, size=n)

    z, p = z_test_sample(data, popmean, popstd, alpha=0.05, alternative='two-sided')
    print(f"Sample mean = {data.mean():.2f}, Population mean = {popmean}")
    print(f"Zâ€‘statistic = {z:.3f}")
    print(f"Pâ€‘value = {p:.4f}")

    # Interpretation
    if p < 0.05:
        print("â†’ Reject H0: sample mean is significantly different from population mean.")
    else:
        print("â†’ Fail to reject H0: no significant difference detected.")
```

```python
# 2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python?
import numpy as np
from scipy import stats

def perform_one_sample_t_test(sample_size=30, population_mean=50, true_mean=52, alpha=0.05):


    simulated_data = np.random.normal(loc=true_mean, scale=10, size=sample_size)

    print(f"Simulated Sample Data (first 5 values): {simulated_data[:5]}")
    print(f"Sample Mean: {np.mean(simulated_data):.2f}")
    print(f"Hypothesized Population Mean (Null Hypothesis): {population_mean}")

    # Perform one-sample t-test
    # The ttest_1samp function returns the t-statistic and the two-tailed p-value.
    t_statistic, p_value = stats.ttest_1samp(simulated_data, population_mean)

    print(f"\nT-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Significance Level (alpha): {alpha}")

    # Make a decision based on the p-value
    if p_value < alpha:
        print("\nConclusion: Reject the null hypothesis.")
        print("There is sufficient evidence to suggest that the sample mean is significantly different from the hypothesized population mean.")
    else:
        print("\nConclusion: Fail to reject the null hypothesis.")
        print("There is not enough evidence to suggest that the sample mean is significantly different from the hypothesized population mean.")

# Example usage of the function
perform_one_sample_t_test()
print("\n--- Another example with a larger difference ---")
perform_one_sample_t_test(true_mean=60)
```

```python
# 3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean?
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std_dev, alpha=0.05, alternative='two-sided'):

    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Calculate the standard error of the mean
    standard_error = population_std_dev / np.sqrt(n)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / standard_error

    # Calculate the p-value based on the alternative hypothesis
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
    elif alternative == 'larger':  # Right-tailed test
        p_value = 1 - norm.cdf(z_statistic)
    elif alternative == 'smaller':  # Left-tailed test
        p_value = norm.cdf(z_statistic)
    else:
        raise ValueError("Invalid 'alternative' argument. Must be 'two-sided', 'larger', or 'smaller'.")

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Sample Size (n): {n}")
    print(f"Z-Statistic: {z_statistic:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis (p-value < {alpha}).")
    else:
        print(f"Fail to reject the null hypothesis (p-value >= {alpha}).")

    return z_statistic, p_value

if __name__ == "__main__":


    # Known population parameters
    known_population_mean = 170  # cm
    known_population_std_dev = 5  # cm

    # Sample data
    sample_heights = [172, 168, 175, 170, 173, 169, 171, 176, 167, 174,
                      170, 172, 169, 175, 171, 168, 173, 170, 174, 166,
                      171, 170, 173, 169, 172, 175, 168, 170, 174, 171] # n=30

    print("--- Two-Sided Z-Test ---")
    z_stat_two_sided, p_value_two_sided = one_sample_z_test(
        sample_data=sample_heights,
        population_mean=known_population_mean,
        population_std_dev=known_population_std_dev,
        alpha=0.05,
        alternative='two-sided'
    )

    print("\n--- Right-Tailed Z-Test (Is sample mean significantly LARGER?) ---")
    z_stat_larger, p_value_larger = one_sample_z_test(
        sample_data=sample_heights,
        population_mean=known_population_mean,
        population_std_dev=known_population_std_dev,
        alpha=0.05,
        alternative='larger'
    )

    print("\n--- Left-Tailed Z-Test (Is sample mean significantly SMALLER?) ---")
    z_stat_smaller, p_value_smaller = one_sample_z_test(
        sample_data=sample_heights,
        population_mean=known_population_mean,
        population_std_dev=known_population_std_dev,
        alpha=0.05,
        alternative='smaller'
    )
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def simulate_and_plot_errors(mu0=0, mu1=1, sigma=1.0, n=30, alpha=0.05, runs=10000):

    # Thresholds
    zcrit = norm.ppf(1 - alpha/2)

    # Simulate Z-statistic distributions under Hâ‚€ and Hâ‚
    zs_h0 = np.random.normal(mu0, sigma/np.sqrt(n), size=runs)
    zs_h1 = np.random.normal(mu1, sigma/np.sqrt(n), size=runs)

    # Compute error rates
    type1 = np.mean(np.abs((zs_h0 - mu0)/(sigma/np.sqrt(n))) > zcrit)
    type2 = np.mean(np.abs((zs_h1 - mu0)/(sigma/np.sqrt(n))) <= zcrit)
    power = 1 - type2

    # Plotting distributions and error regions
    x = np.linspace(min(mu0, mu1) - 4*sigma/np.sqrt(n),
                    max(mu0, mu1) + 4*sigma/np.sqrt(n), 1000)
    plt.figure(figsize=(10,6))
    plt.plot(x, norm.pdf(x, mu0, sigma/np.sqrt(n)), label=f'Hâ‚€: Î¼={mu0}')
    plt.plot(x, norm.pdf(x, mu1, sigma/np.sqrt(n)), label=f'Hâ‚: Î¼={mu1}')

    # Shade Type I regions
    plt.fill_between(x, norm.pdf(x, mu0, sigma/np.sqrt(n)),
                     where=np.abs((x - mu0)/(sigma/np.sqrt(n))) > zcrit,
                     color='red', alpha=0.3, label=f'Type I error (Î± â‰ˆ {type1:.3f})')
    # Shade Type II region
    plt.fill_between(x, norm.pdf(x, mu1, sigma/np.sqrt(n)),
                     where=np.abs((x - mu0)/(sigma/np.sqrt(n))) <= zcrit,
                     color='blue', alpha=0.3, label=f'Type II error (Î² â‰ˆ {type2:.3f})')

    # Critical lines
    plt.axvline(mu0 + zcrit*sigma/np.sqrt(n), color='black', linestyle='--')
    plt.axvline(mu0 - zcrit*sigma/np.sqrt(n), color='black', linestyle='--')

    plt.title('Type I & II Error Visualization for Two-Tailed Z-Test')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

    return type1, type2, power

# Example usage
if __name__ == "__main__":
    Î± = 0.05
    t1, t2, pw = simulate_and_plot_errors(mu0=100, mu1=103, sigma=15, n=36, alpha=Î±, runs=20000)
    print(f"Estimated Type I error (Î±): {t1:.3f}")
    print(f"Estimated Type II error (Î²): {t2:.3f}")
    print(f"Statistical Power (1 âˆ’ Î²): {pw:.3f}")
```

```python
# 6.  Write a Python program to perform an independent T-test and interpret the results?

import numpy as np
from scipy import stats

def perform_independent_ttest(group1_data, group2_data, equal_variances=True):

    t_statistic, p_value = stats.ttest_ind(group1_data, group2_data, equal_var=equal_variances)

    interpretation = f"Independent Samples T-Test Results:\n"
    interpretation += f"  T-statistic: {t_statistic:.3f}\n"
    interpretation += f"  P-value: {p_value:.3f}\n"

    alpha = 0.05  # Significance level

    if p_value < alpha:
        interpretation += f"  Since the p-value ({p_value:.3f}) is less than the significance level ({alpha}),\n"
        interpretation += "  we reject the null hypothesis. There is a statistically significant difference between the means of the two groups."
    else:
        interpretation += f"  Since the p-value ({p_value:.3f}) is greater than the significance level ({alpha}),\n"
        interpretation += "  we fail to reject the null hypothesis. There is no statistically significant difference between the means of the two groups."

    return t_statistic, p_value, interpretation

# Example Usage:
# Sample data for two independent groups
group_a = [22, 25, 20, 28, 23, 26, 21, 24]
group_b = [18, 19, 17, 20, 16, 21, 19, 18]

# Perform the t-test assuming equal variances
t_stat_equal_var, p_val_equal_var, interpretation_equal_var = perform_independent_ttest(group_a, group_b, equal_variances=True)
print("\n--- Assuming Equal Variances (Student's t-test) ---")
print(interpretation_equal_var)

# Perform the t-test assuming unequal variances (Welch's t-test)
t_stat_unequal_var, p_val_unequal_var, interpretation_unequal_var = perform_independent_ttest(group_a, group_b, equal_variances=False)
print("\n--- Assuming Unequal Variances (Welch's t-test) ---")
print(interpretation_unequal_var)
```

```python
# 8. Simulate data and perform both Z-test and T-test, then compare the results using Python?
import numpy as np
from scipy import stats

def perform_tests(sample_size, pop_mean, pop_std, alpha=0.05):


    # Simulate data from a normal distribution
    np.random.seed(42)  # for reproducibility
    sample_data = np.random.normal(loc=pop_mean, scale=pop_std, size=sample_size)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1) # ddof=1 for sample standard deviation

    print(f"--- Simulation with Sample Size: {sample_size} ---")
    print(f"Simulated Sample Mean: {sample_mean:.2f}")
    print(f"Simulated Sample Standard Deviation: {sample_std:.2f}")

    # --- Z-test ---
    # Z-test assumes known population standard deviation
    z_statistic = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
    p_value_z = 2 * (1 - stats.norm.cdf(abs(z_statistic))) # Two-tailed p-value

    print("\n--- Z-test Results ---")
    print(f"Z-statistic: {z_statistic:.2f}")
    print(f"P-value (Z-test): {p_value_z:.4f}")
    if p_value_z < alpha:
        print(f"Result: Reject the null hypothesis (at alpha={alpha}) - Sample mean is significantly different from population mean.")
    else:
        print(f"Result: Fail to reject the null hypothesis (at alpha={alpha}) - No significant difference found.")

    # --- T-test ---
    # T-test uses sample standard deviation and is suitable for unknown population std or smaller samples
    t_statistic, p_value_t = stats.ttest_1samp(sample_data, pop_mean)

    print("\n--- T-test Results ---")
    print(f"T-statistic: {t_statistic:.2f}")
    print(f"P-value (T-test): {p_value_t:.4f}")
    if p_value_t < alpha:
        print(f"Result: Reject the null hypothesis (at alpha={alpha}) - Sample mean is significantly different from population mean.")
    else:
        print(f"Result: Fail to reject the null hypothesis (at alpha={alpha}) - No significant difference found.")

    # --- Comparison ---
    print("\n--- Comparison ---")
    if p_value_z < alpha and p_value_t < alpha:
        print("Both Z-test and T-test suggest rejecting the null hypothesis.")
    elif p_value_z >= alpha and p_value_t >= alpha:
        print("Both Z-test and T-test suggest failing to reject the null hypothesis.")
    elif p_value_z < alpha and p_value_t >= alpha:
        print("Z-test suggests rejecting, while T-test suggests failing to reject (likely due to sample size/std assumption).")
    else: # p_value_z >= alpha and p_value_t < alpha
        print("T-test suggests rejecting, while Z-test suggests failing to reject (less common, but possible).")

# Example Usage
perform_tests(sample_size=30, pop_mean=100, pop_std=10)
print("\n" + "="*50 + "\n")
perform_tests(sample_size=15, pop_mean=100, pop_std=10)
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance?
import numpy as np
import scipy.stats as stats

def mean_confidence_interval(data, confidence=0.95):

    a = np.array(data)
    n = len(a)
    mean = np.mean(a)
    sem = stats.sem(a)  # standard error of the mean
    # critical t-value for two-tailed test
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)
    margin = sem * t_crit

    return mean, mean - margin, mean + margin


# Example usage:
if __name__ == "__main__":
    sample = [12, 15, 14, 10, 13, 17, 14, 15, 16, 14]
    m, lower, upper = mean_confidence_interval(sample, 0.95)
    print(f"Mean = {m:.2f}")
    print(f"95% CI = [{lower:.2f}, {upper:.2f}]")
```

Explanation and Significance:

The function confidence_interval calculates the confidence interval for the population mean based on a sample. Here's a breakdown of its components and their significance:

1. Data Input:
The function takes sample data as a NumPy array (or any array-like object) and a confidence level as input. The confidence parameter defaults to 95% (0.95).

2. Sample Mean and Standard Error:
The function first calculates the sample mean and the standard error of the mean (SEM). The SEM is a measure of the variability of the sample mean and is calculated as the standard deviation of the sample divided by the square root of the sample size.

3. Degrees of Freedom:
For smaller sample sizes (typically n <= 30), the t-distribution is used. The degrees of freedom (df) are calculated as n-1, where n is the sample size.

4. Critical Value (t or z):
If the sample size is small (n <= 30), the function calculates the critical t-value using st.t.ppf((1 + confidence) / 2., n-1). The ppf function calculates the percent point function (inverse of the cumulative distribution function) for the t-distribution.

If the sample size is large (n > 30), the function uses the z-distribution and calculates the critical z-value using st.norm.ppf((1 + confidence) / 2).

5. Margin of Error:
The margin of error is calculated by multiplying the critical value (t or z) by the standard error. This represents the amount added and subtracted from the sample mean to create the confidence interval.

6. Confidence Interval:
Finally, the confidence interval is constructed by subtracting the margin of error from the sample mean to get the lower bound and adding it to the sample mean to get the upper bound.

 - Significance of Confidence Intervals:
Estimating Population Parameter:
Confidence intervals provide a range of plausible values for the population mean, given the sample data. Instead of relying on a single point estimate (the sample mean), a confidence interval gives a more informative picture of the potential range of the true population mean.

 - Level of Confidence:
The confidence level (e.g., 95%) indicates the probability that the true population mean falls within the calculated interval. A 95% confidence interval means that if we were to repeat the sampling process many times and calculate a confidence interval each time, 95% of those intervals would contain the true population mean.

 - Precision and Uncertainty:
The width of the confidence interval reflects the precision of the estimate. A narrower interval suggests a more precise estimate (less uncertainty), while a wider interval indicates greater uncertainty.

 - Decision Making:
Confidence intervals are useful in various applications, including hypothesis testing, quality control, and research studies, where understanding the uncertainty associated with estimates is crucial.

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data?


import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(sample_data, confidence_level=0.95):

    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(sample_data)

    # Calculate the t-critical value
    # For a two-tailed interval, we need (1 + confidence_level) / 2
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    # Calculate the standard error of the mean
    standard_error = sample_std / np.sqrt(sample_size)

    # Calculate the margin of error
    margin_of_error = t_critical * standard_error

    return margin_of_error

# Example usage:
if __name__ == "__main__":
    # Sample data (e.g., heights of students)
    data = [165, 170, 172, 168, 175, 169, 171, 173, 167, 170]

    # Calculate margin of error for 95% confidence
    moe_95 = calculate_margin_of_error(data, confidence_level=0.95)
    print(f"Margin of error (95% confidence): {moe_95:.2f}")

    # Calculate margin of error for 99% confidence
    moe_99 = calculate_margin_of_error(data, confidence_level=0.99)
    print(f"Margin of error (99% confidence): {moe_99:.2f}")

    # Calculate confidence interval
    sample_mean = np.mean(data)
    lower_ci_95 = sample_mean - moe_95
    upper_ci_95 = sample_mean + moe_95
    print(f"95% Confidence Interval: ({lower_ci_95:.2f}, {upper_ci_95:.2f})")
```

```python
# 11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?
def bayes_theorem(p_h, p_e_given_h, p_e_given_not_h):

    # Calculate P(not H)
    p_not_h = 1 - p_h

    # Calculate P(E) using the law of total probability
    p_e = (p_e_given_h * p_h) + (p_e_given_not_h * p_not_h)

    # Calculate P(H|E) using Bayes' Theorem
    if p_e == 0:
        return 0  # Avoid division by zero if P(E) is zero
    p_h_given_e = (p_e_given_h * p_h) / p_e
    return p_h_given_e

# Define probabilities for a disease diagnosis example
prevalence_of_disease = 0.01  # P(Disease) - Prior
test_sensitivity = 0.95       # P(Positive Test | Disease) - Likelihood
test_specificity = 0.90       # P(Negative Test | No Disease)

# Calculate P(Positive Test | No Disease) from specificity
p_positive_test_given_no_disease = 1 - test_specificity

# Calculate the posterior probability
posterior_prob = bayes_theorem(
    prevalence_of_disease,
    test_sensitivity,
    p_positive_test_given_no_disease
)

print(f"Prior probability of disease: {prevalence_of_disease:.4f}")
print(f"Likelihood of positive test given disease: {test_sensitivity:.4f}")
print(f"Likelihood of positive test given no disease: {p_positive_test_given_no_disease:.4f}")
print(f"Posterior probability of disease given a positive test: {posterior_prob:.4f}")
```

Explanation of the Process:Â     

  - Define Prior Probability (\(P(H)\)): This is the initial belief in the hypothesis before observing any evidence. In the example, it is the prevalence of the disease in the population.
  
  - Define Likelihood (\(P(E|H)\)): This is the probability of observing the evidence given that the hypothesis is true. In the example, it is the sensitivity of the test (probability of a positive test given the disease is present).
   
   
  - Define Likelihood of Evidence given the Hypothesis is False (\(P(E|\neg H)\)): This is the probability of observing the evidence given that the hypothesis is false. In the example, it is the probability of a positive test given no disease, derived from the test's specificity.

  - Calculate Marginal Likelihood (\(P(E)\)): This represents the overall probability of observing the evidence, regardless of whether the hypothesis is true or false. It is calculated using the law of total probability: \(P(E)=P(E|H)P(H)+P(E|\neg H)P(\neg H)\).

 - Calculate Posterior Probability (\(P(H|E)\)): Bayes' Theorem is applied to combine the prior belief with the observed evidence and its likelihoods to obtain the updated probability of the hypothesis. This posterior probability reflects the refined belief after considering the new information.

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python?

import pandas as pd
from scipy.stats import chi2_contingency

def perform_chi_square_test(data, var1, var2):

    # Create a contingency table (cross-tabulation)
    contingency_table = pd.crosstab(data[var1], data[var2])

    # Perform the Chi-square test
    chi2_stat, p_value, dof, expected_freq = chi2_contingency(contingency_table)

    return chi2_stat, p_value, dof, expected_freq

# Example Usage:
if __name__ == "__main__":
    # Sample data
    data = {
        'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
        'Preference': ['A', 'B', 'A', 'A', 'B', 'B', 'A', 'B']
    }
    df = pd.DataFrame(data)

    print("Contingency Table:")
    print(pd.crosstab(df['Gender'], df['Preference']))

    # Perform the Chi-square test
    chi2_stat, p_value, dof, expected_freq = perform_chi_square_test(df, 'Gender', 'Preference')

    print(f"\nChi-square Statistic: {chi2_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Degrees of Freedom: {dof}")
    print("Expected Frequencies:")
    print(expected_freq)

    # Interpret the results (common significance level alpha = 0.05)
    alpha = 0.05
    if p_value < alpha:
        print("\nResult: Reject the null hypothesis. There is a significant association between Gender and Preference.")
    else:
        print("\nResult: Fail to reject the null hypothesis. There is no significant association between Gender and Preference.")
```

```python
# 13.  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data?

import numpy as np

def calculate_expected_frequencies(observed_data):

    row_totals = np.sum(observed_data, axis=1, keepdims=True)
    col_totals = np.sum(observed_data, axis=0, keepdims=True)
    grand_total = np.sum(observed_data)

    expected_frequencies = (row_totals * col_totals) / grand_total
    return expected_frequencies

if __name__ == "__main__":
    # Example observed data (contingency table)
    # Rows represent one categorical variable, columns represent another.
    observed = np.array([
        [10, 20, 30],
        [15, 25, 35],
        [5, 10, 15]
    ])

    expected = calculate_expected_frequencies(observed)

    print("Observed Frequencies:")
    print(observed)
    print("\nExpected Frequencies:")
    print(expected)
```

```python
# 14.  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution?
import numpy as np
from scipy.stats import chisquare
import pandas as pd

def chi_square_gof(observed, expected=None, expected_probs=None, alpha=0.05):

    obs = np.array(observed)
    if expected_probs is not None:
        expected = np.array(expected_probs) * obs.sum()
    elif expected is None:
        expected = np.ones_like(obs) * (obs.sum() / len(obs))

    stat, p = chisquare(f_obs=obs, f_exp=expected)
    dof = len(obs) - 1
    conclusion = ("Reject Hâ‚€, distributions differ"
                  if p < alpha else
                  "Fail to reject Hâ‚€, no evidence of difference")

    return {
        'chi2_stat': stat,
        'p_value': p,
        'dof': dof,
        'expected': expected,
        'conclusion': conclusion
    }

if __name__ == "__main__":
    # Example 1: Goodness-of-fit to a fair die (uniform)
    observed_die = [8, 9, 10, 11, 12, 10]  # counts for faces 1â€“6
    result_die = chi_square_gof(observed_die, expected_probs=[1/6]*6)
    print("Fair die test:\n", pd.Series(result_die))

    # Example 2: Fit to a custom expected distribution
    observed = [30, 50, 20]
    expected = [25, 50, 25]
    result_custom = chi_square_gof(observed, expected=expected)
    print("\nCustom distribution test:\n", pd.Series(result_custom))
```
