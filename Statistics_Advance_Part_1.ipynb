4. What are probability distribution functions (PDF)?


Ans:A Probability Density Function (PDF) describes the likelihood of a continuous random variable falling within a specific range of values. It's a function that represents the density of probability rather than directly giving probabilities.

Key characteristics of a PDF:

 - Defined for continuous variables:

PDFs are used for random variables that can take any value within a given range (e.g., height, weight).

 - Probability as area:

The probability of a continuous random variable falling within a certain interval is given by the area under the PDF curve within that interval.

 - Non-negative:

The PDF is always greater than or equal to zero for all possible values of the random variable.

 - Total area is one:

The integral of the PDF over its entire range is equal to 1, representing the certainty of the variable taking some value within its range.

 - In simpler terms: Imagine a curve representing the distribution of heights in a population. The PDF would show how "dense" the heights are at different values. A taller section of the curve means there are more people with heights in that range. The probability of a randomly selected person having a height within a specific interval is the area under that part of the curve.

It's important to distinguish a PDF from a Probability Mass Function (PMF). A PMF is used for discrete random variables (variables that can only take specific, separate values, like the number of heads in coin flips). A PDF is specifically for continuous random variables.

The PDF gives the probability density at a specific point, while the CDF gives the probability that a random variable is less than or equal to a specific value.

1. PDF:

 - Describes the likelihood of a random variable taking on a specific value.

 - For continuous variables, it's the probability density, not a direct probability at a single point.

 - The area under the PDF curve over an interval represents the probability of the variable falling within that interval.

 - Can be integrated to obtain the CDF.

2. CDF:

 - Gives the probability that a random variable is less than or equal to a specific value.

 - Ranges from 0 to 1, representing the cumulative probability.

 - Is always non-decreasing.

 - Can be obtained by integrating the PDF.

 Relationship:

 - The CDF is the integral of the PDF. Conversely, the PDF is the derivative of the CDF.

In simpler terms: Imagine rolling a die. The PDF would show the probability of rolling each specific number (e.g., 1/6 for any number). The CDF would show the probability of rolling a number less than or equal to a specific value (e.g., the probability of rolling a 2 or less is 2/6).

8.  What is the binomial distribution, and how is it used in probability?


Ans:The binomial distribution is a probability distribution that models the probability of obtaining a specific number of successes in a fixed number of independent trials, where each trial has only two possible outcomes (success or failure) and the probability of success is constant across all trials. It's used in probability to analyze situations with binary outcomes, like coin flips, defect rates, or survey responses.

 - Key Concepts:

  - Independent Trials: The outcome of one trial does not affect the outcome of other trials.

  - Fixed Number of Trials (n): The binomial distribution requires a predetermined number of trials.

  - Two Possible Outcomes: Each trial results in either success or failure.

  - Constant Probability of Success (p): The probability of success remains the same for every trial.

 - Formula:

  - The probability of exactly k successes in n trials is given by:

  - P(X = k) = (n choose k) * p^k * (1-p)^(n-k)

 - Where:

  - P(X = k) is the probability of exactly k successes

  - (n choose k) is the binomial coefficient, representing the number of ways to choose k successes from n trials

  - p is the probability of success on a single trial

   - (1-p) is the probability of failure

  - n is the total number of trials

  - k is the number of successful trials

 - Example:

  - Imagine flipping a fair coin 5 times. What's the probability of getting exactly 3 heads?

  - n = 5 (number of trials)

  - k = 3 (number of successes, i.e., heads)

  - p = 0.5 (probability of getting heads on a single flip)

 - Using the formula:

   - P(X = 3) = (5 choose 3) * (0.5)^3 * (0.5)^2 = 10 * 0.125 * 0.25 = 0.3125

  - So, the probability of getting exactly 3 heads in 5 flips is 0.3125 or 31.25%.

Applications:

 - Quality Control:
Determining the probability of finding a certain number of defective items in a production batch.

 - Marketing:
Analyzing the likelihood of a customer making a purchase after receiving a promotional email.

 - Medical Research:
Assessing the success rate of a new drug.

 - Social Sciences:
Modeling the proportion of people who hold a particular opinion in a survey.

 - Symmetry:

  - Bell Shape:
The distribution's shape resembles a bell, with the highest point at the mean. The curve gradually tapers off on both sides as you move away from the mean


 - Unimodality:
The distribution has a single peak (mode) at the mean.

 - Asymptotic Tails:
The tails of the curve extend infinitely in both directions, approaching the x-axis but never touching it.

 - Mean, Median, and Mode:
In a normal distribution, the mean (average), median (middle value), and mode (most frequent value) are all equal and located at the center of the curve.

 - Standard Deviation:
The standard deviation determines the spread or width of the curve. A smaller standard deviation results in a narrower, taller curve, while a larger standard deviation results in a wider, flatter curve.

 - Total Area Under the Curve:
The total area under the normal distribution curve is equal to 1 (or 100%).

 - Empirical Rule:
Approximately 68% of the data falls within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Ans:The Central Limit Theorem (CLT) states that the distribution of sample means will approximate a normal distribution, regardless of the original population's distribution, as the sample size becomes sufficiently large. This is crucial because it allows statisticians to apply statistical methods designed for normal distributions to a wide range of problems, even when the underlying data isn't normally distributed.

Key aspects of the CLT:

 - Sample Means:
The theorem focuses on the distribution of sample means, not individual data points.

 - Large Sample Size:
The approximation to a normal distribution improves as the sample size increases. A common rule of thumb is that a sample size of 30 or more is generally considered large enough.

 - Independence and Identical Distribution:
The samples should be independent and identically distributed (i.i.d.) for the CLT to hold strictly.

Why is the CLT critical in statistics:

 - Foundation for Inference:
Many statistical methods, like hypothesis testing and confidence interval construction, rely on the assumption of a normal distribution.

 - Simplifies Analysis:
By allowing the use of normal distribution-based techniques, the CLT simplifies the analysis of data from various distributions.

 - Generalizability:
The CLT enables us to make inferences about a population based on a sample, even if the population isn't normally distributed.

 - Applications in Machine Learning:
The CLT is also fundamental in machine learning, particularly for validating models and understanding the behavior of algorithms.




Why is the CLT so important:

 - Universal Normality for Statistics:
It lets us treat the sample mean as normally distributed, even if the underlying data aren’t normal. This underpins inference techniques like confidence intervals and hypothesis tests


 - Simplifies Complex Distributions:
You can aggregate various data types—skewed, discrete, heavy‑tailed—and still rely on normality for analysis

 - Foundation for Modern Statistics & ML:
Core results (e.g., asymptotic normality of estimators like MLEs) are grounded in the CLT

 - Handles Uncertainty via Quantitative Bounds:
The Berry–Esseen theorem refines CLT, giving error bounds and convergence rates (approximately 1/√n)




Key Conditions for the CLT to Hold:

 - Independence: Observations do not influence each other.

 - Identical Distribution (i.i.d.): Each has the same μ and σ².

 - Finite variance: Distributions with infinite variance (e.g., certain Pareto) do not satisfy CLT .

 - Sufficiently large sample size: Often n ≥ 30 is a good rule of thumb .

14. How does the Central Limit Theorem relate to the normal distribution?


Ans:The central limit theorem (CLT) essentially states that the distribution of sample means will approximate a normal distribution, regardless of the underlying population distribution, as long as the sample size is sufficiently large. This means that even if you are sampling from a population that is not normally distributed, the averages of multiple samples will tend towards a normal distribution.

Here's a breakdown:

 - Normal Distribution:
The normal distribution, also known as the Gaussian distribution, is a bell-shaped probability distribution characterized by its mean and standard deviation.

 - Central Limit Theorem:
The CLT focuses on the distribution of sample means. It states that if you take many random samples from a population and calculate the mean of each sample, the distribution of these sample means will be approximately normal, even if the original population isn't normally distributed.

 - Relationship:
The CLT's significance lies in its connection to the normal distribution. It allows us to use the properties of the normal distribution (like probabilities associated with specific ranges) to analyze and make inferences about sample means, even when the population distribution is unknown or non-normal.

 - Sample Size:
The larger the sample size, the better the approximation of the sample means to a normal distribution. A common rule of thumb is that a sample size of 30 or more is generally considered "sufficiently large" for the CLT to apply, but this can vary depending on the shape of the original population distribution.

 - Example:
Imagine a population with a highly skewed distribution, like the distribution of income. If you take many random samples from this population and calculate the average income for each sample, the distribution of these average incomes will tend towards a normal distribution as the sample size increases.

 - Central Limit Theorem Example:
The CLT provides a powerful tool for statistical analysis by allowing us to leverage the properties of the normal distribution when dealing with sample means, even when the population itself is not normally distributed.




Without the CLT, statistical analysis would be severely limited to data known to be normally distributed. Instead, the CLT lets us:

 - Use z-tests and confidence intervals broadly

 - Approximate distributions of sums and averages as normal

 - Rely on normal-based methods in regression, polls, quality control, and more—even when raw data deviate from normality.

Ans:Z-statistics, calculated using a z-test, are applied in hypothesis testing to determine if there's a significant difference between a sample mean and a population mean, or between the means of two samples, when certain conditions are met. Specifically, z-tests are used when the population variance is known, or when the sample size is large (typically n ≥ 30). The z-statistic helps assess the likelihood of observing the sample data if the null hypothesis is true.

Here's a more detailed breakdown:

1. When to use z-tests:

 - Large sample sizes:
Z-tests are most reliable when dealing with larger sample sizes, as the sampling distribution of the sample mean approaches a normal distribution (Central Limit Theorem).

 - Known population variance:
If the population standard deviation (and thus the variance) is known, a z-test can be used.

 - Comparing sample means to a population mean:
Z-tests can determine if a sample mean is significantly different from a known population mean.

 - Comparing two sample means:
Z-tests can also be used to compare the means of two independent samples.

2. Steps in using z-tests for hypothesis testing:

  - State the null and alternative hypotheses:
Define what you're trying to prove or disprove. The null hypothesis (H0) is a statement of no effect or no difference, while the alternative hypothesis (Ha) is the statement you're trying to find evidence for.

  - Set the significance level (alpha):
This determines the threshold for rejecting the null hypothesis. Common values are 0.05 or 0.01.

  - Calculate the z-statistic:
This is done using a formula that compares the sample mean(s) to the population mean(s) and considers the standard deviation and sample size.

  - Determine the p-value or critical value:
The p-value is the probability of observing the data (or more extreme data) if the null hypothesis is true. The critical value is a threshold based on the significance level.

  - Make a decision:
If the p-value is less than or equal to the significance level (or if the z-statistic falls in the rejection region), reject the null hypothesis. Otherwise, fail to reject the null hypothesis.

3. Types of z-tests:

  - One-sample z-test: Compares a sample mean to a population mean.

  - Two-sample z-test: Compares the means of two independent samples.

  - Paired z-test: A type of two-sample test where the samples are related (e.g., before-and-after measurements on the same subjects).




 - Applications in various fields:

  - Quality Control: Monitoring production processes to ensure products meet specified standards.

  - Finance: Analyzing stock market data and making investment decisions.

  - Healthcare: Comparing the effectiveness of different treatments or medications.

  - Social Sciences: Comparing the opinions or behaviors of different groups.

  - Marketing: Evaluating the effectiveness of marketing campaigns.


 - Key Considerations:

  - Assumptions:
Z-tests assume that the data is normally distributed and that the population standard deviation is known (or the sample size is large).

  - Sample Size:
Z-tests are generally more appropriate for larger sample sizes (n ≥ 30). For smaller samples, t-tests are often preferred.

You can then use a one-sample z-test to compare the school's average height to the national average height.

 - A point estimate is a single value calculated from sample data that is used to estimate a population parameter.

 - For example, the sample mean is a point estimate of the population mean.

 - Point estimates are straightforward to calculate and understand, but they do not provide any information about the precision or reliability of the estimate.

 - A point estimate alone doesn't tell you how close it is to the true population value.

 - An interval estimate provides a range of values within which the population parameter is likely to fall, along with a specified level of confidence.

 - A common type of interval estimate is the confidence interval, which gives a range of values that is likely to contain the true population parameter.

 - Interval estimates are more informative than point estimates because they provide information about the uncertainty associated with the estimate.

 - For instance, a 95% confidence interval means that if you were to repeat the sampling process many times, 95% of the calculated confidence intervals would contain the true population parameter.

 - Interval estimates are often preferred over point estimates because they provide a more complete picture of the estimation process by quantifying the uncertainty.

19. What is the relationship between a Z-score and a confidence interval?

Ans:A Z-score and a confidence interval are related in that the Z-score is used to calculate the margin of error, which is then used to construct the confidence interval. Specifically, the Z-score represents the number of standard deviations a data point is from the mean in a normal distribution. In confidence interval calculations, a specific Z-score (determined by the desired confidence level) is used to define the range around a sample mean within which the true population mean is likely to fall.

Here's a more detailed explanation:

 - Z-score:
The Z-score measures how many standard deviations a data point is away from the mean of a dataset. It's calculated using the formula: z = (x - μ) / σ, where x is the data point, μ is the population mean, and σ is the population standard deviation.

 - Confidence Interval:
A confidence interval provides a range of values within which a population parameter (like the mean) is likely to fall with a certain level of confidence. For example, a 95% confidence interval means that if you were to repeat the sampling process many times, 95% of the calculated intervals would contain the true population mean.

 - Relationship:
To calculate a confidence interval for the population mean, you need to determine the margin of error. The margin of error is calculated by multiplying the Z-score (corresponding to your desired confidence level) by the standard error of the mean. The standard error of the mean is calculated by dividing the population standard deviation by the square root of the sample size.

 - Example:
If you want to calculate a 95% confidence interval for the population mean, you would first find the Z-score for 95% confidence, which is 1.96 according to Z-score tables. Then, you would use this Z-score, along with the sample mean, population standard deviation, and sample size, to calculate the margin of error and construct the confidence interval.

The Z-score acts as a bridge between the desired level of confidence and the width of the confidence interval. A higher confidence level (e.g., 99%) will correspond to a larger Z-score, resulting in a wider confidence interval, while a lower confidence level (e.g., 90%) will have a smaller Z-score and a narrower interval.




Formula for Confidence Interval Using Z-Score
To calculate a confidence interval for a population mean when the standard deviation is known:

 - CI = x̄ ± z * (σ / √n)

Where:

  -  x̄ = sample mean

  - z = Z-score corresponding to the desired confidence level

  - σ = population standard deviation

  - n = sample size

This formula provides the range within which the true population mean is expected to lie with the specified confidence level.


 - Z-scores are critical values that determine the width of confidence intervals.

 - Higher confidence levels require larger Z-scores, resulting in wider intervals.

 - Confidence intervals provide a range that likely contains the true population parameter, with the Z-score ensuring the interval aligns with the desired confidence level

21.  What are the assumptions for applying the Central Limit Theorem?

Ans:The Central Limit Theorem (CLT) requires a few key assumptions to be met for the sampling distribution of sample means to approximate a normal distribution. These include: random sampling, independence of samples, a sufficiently large sample size, and sometimes, that the sample size is less than 10% of the population when sampling without replacement.

Here's a more detailed explanation:

1. Random Sampling:
The sample must be drawn randomly from the population. This ensures that each member of the population has an equal chance of being selected, which is crucial for reducing bias and making the sample representative of the population.

2. Independence of Samples:
The individual observations within the sample should be independent of each other. This means that the selection of one data point should not influence the selection or value of any other data point in the sample.

3. Sufficiently Large Sample Size:
The sample size needs to be large enough for the CLT to be effective. While a rule of thumb is often cited as n ≥ 30, the necessary sample size can vary depending on the shape of the population distribution. For distributions that are heavily skewed or have outliers, a larger sample size might be needed.

4. Sampling Without Replacement:
If sampling is done without replacement, the sample size should not exceed 10% of the total population. This helps to maintain the independence of the samples.

These assumptions ensure that the sampling distribution of sample means will converge towards a normal distribution as the sample size increases, regardless of the original population's distribution (as long as it has a finite variance).

Putting It All Together:

If your data meet the following criteria:

 - Random + Independent

 - (Approximately) Identical Distribution

 - Finite Mean & Variance

 - Large Sample Size

Then the sampling distribution of the mean will be approximately normal, centered at μ, with standard deviation σ/√n

22.  What is the concept of expected value in a probability distribution?

Ans:The expected value in a probability distribution represents the long-run average outcome of a random variable. It's a weighted average of all possible values, where each value is weighted by its corresponding probability. Here's a more detailed explanation:

 -  Probability Distribution:
A probability distribution describes the likelihood of each possible outcome of a random variable.

 -  Random Variable:
A random variable is a variable whose value is a numerical outcome of a random phenomenon.

 - Expected Value (E(X)):
For a discrete random variable, the expected value is calculated by multiplying each possible outcome by its probability and summing the results. The formula is: E(X) = Σ [x * P(x)], where 'x' represents each outcome and 'P(x)' its probability.

 - Interpreting Expected Value:
The expected value is not necessarily a value that will be observed in any single trial of the experiment. Instead, it represents the average outcome over many repetitions. If you were to repeatedly run the experiment, the average of the outcomes would tend towards the expected value.

 - Example: Imagine flipping a fair coin. The possible outcomes are heads (H) and tails (T), each with a probability of 0.5. The expected value of the number of heads in a single flip is (1 * 0.5) + (0 * 0.5) = 0.5. This means that, on average, you would expect to get half a head (which is, of course, impossible in a single flip) if you flipped the coin many times.

In essence, the expected value provides a way to quantify the central tendency or average outcome of a probability distribution, helping in decision-making and understanding the long-term behavior of random events.

```python
# 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?

import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(x, a, b):

    if not isinstance(x, np.ndarray):
        x = np.array([x])

    cdf_values = np.zeros_like(x, dtype=float)
    n = b - a + 1  # Number of possible outcomes

    for i, val in enumerate(x):
        if val < a:
            cdf_values[i] = 0
        elif val >= b:
            cdf_values[i] = 1
        else:
            # For values within the range [a, b), the CDF is (floor(x) - a + 1) / n
            cdf_values[i] = (np.floor(val) - a + 1) / n
    return cdf_values

# Define the parameters of the discrete uniform distribution
lower_bound = 1
upper_bound = 6

# Generate a range of x values for plotting
# We include values before, within, and after the distribution range
x_values = np.linspace(lower_bound - 2, upper_bound + 2, 500)

# Calculate the CDF values
cdf_y_values = discrete_uniform_cdf(x_values, lower_bound, upper_bound)

# Plotting the CDF
plt.figure(figsize=(10, 6))
plt.plot(x_values, cdf_y_values, drawstyle='steps-post', label='CDF of Discrete Uniform Distribution')
plt.title(f'Cumulative Distribution Function of Discrete Uniform Distribution (a={lower_bound}, b={upper_bound})')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.yticks(np.arange(0, 1.1, 0.1)) # Set y-axis ticks for better visualization
plt.xticks(np.arange(lower_bound - 2, upper_bound + 3, 1)) # Set x-axis ticks
plt.legend()
plt.show()
```

```python
# 13.Generate random variables and calculate their corresponding probabilities using the binomial distribution?

import math

def binomial_probability(n, k, p):

  combinations = math.comb(n, k) # or math.factorial(n) // (math.factorial(k) * math.factorial(n - k))
  probability = combinations * (p**k) * ((1-p)**(n-k))
  return probability

# Example usage:
n_trials = 5
num_successes = 3
probability_of_success = 0.5

probability = binomial_probability(n_trials, num_successes, probability_of_success)
print(f"The probability of {num_successes} successes in {n_trials} trials is: {probability}")
```

```python
# 13.Generate random variables and calculate their corresponding probabilities using the binomial distribution?


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def simulate_binomial(n, p, size=1000, plot=True):

    # 1️⃣ Generate random samples
    samples = binom.rvs(n=n, p=p, size=size, random_state=42)

    # 2️⃣ Possible outcomes
    k = np.arange(n + 1)

    # 3️⃣ Compute theoretical PMF P(X = k)
    pmf = binom.pmf(k, n, p)

    # 4️⃣ Optional plot
    if plot:
        plt.figure(figsize=(10, 6))
        plt.bar(k - 0.2, pmf, width=0.4, label='Theoretical PMF')

        # Empirical frequencies
        freq = np.bincount(samples, minlength=n + 1) / size
        plt.bar(k + 0.2, freq, width=0.4, alpha=0.7, label='Empirical frequency')

        plt.xlabel('Number of Successes (k)')
        plt.ylabel('Probability')
        plt.title(f'Binomial Distribution (n={n}, p={p})')
        plt.legend()
        plt.grid(True)
        plt.show()

    # 5️⃣ Return results
    return samples, k, pmf

# Example:
n, p, size = 10, 0.4, 5000
samples, k, pmf = simulate_binomial(n, p, size)
```

```python
# 14.  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from statistics import NormalDist

def calculate_z(x, mu, sigma):
    """Compute Z‑score for a data point."""
    return (x - mu) / sigma

def z_to_probability(z):
    """Compute P(Z ≤ z) using standard normal CDF."""
    return norm.cdf(z)

def plot_z(z):
    """Plot standard normal PDF and shade the area to the left of z."""
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Standard Normal PDF')
    plt.fill_between(x[x <= z], y[x <= z], alpha=0.3, label=f'P(Z ≤ {z:.2f})')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title('Standard Normal Distribution')
    plt.xlabel('Z‑score')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    # Example parameters
    x = 650         # data point
    mu, sigma = 500, 100  # mean & standard deviation

    z = calculate_z(x, mu, sigma)
    prob = z_to_probability(z)

    print(f"Z‑score of {x} (μ={mu}, σ={sigma}) is {z:.2f}")
    print(f"P(Z ≤ {z:.2f}) = {prob:.4f} → {(prob*100):.2f}%")

    plot_z(z)
```

```python
# 15.  Implement hypothesis testing using Z-statistics for a sample dataset?

import numpy as np
from scipy import stats

def z_test_one_sample(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):

    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Calculate the p-value based on the alternative hypothesis
    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_statistic)
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_statistic)
    else:
        raise ValueError("Invalid 'alternative' specified. Choose 'two-sided', 'less', or 'greater'.")

    # Make a decision
    if p_value < alpha:
        decision = "Reject the null hypothesis."
    else:
        decision = "Fail to reject the null hypothesis."

    return {
        "z_statistic": z_statistic,
        "p_value": p_value,
        "decision": decision
    }

# Example Usage:
if __name__ == "__main__":
    # Sample data
    sample = [78, 82, 79, 85, 80, 81, 77, 83, 84, 80]

    # Hypothesized population mean
    mu_0 = 80

    # Known population standard deviation
    sigma = 3

    # Significance level
    alpha_level = 0.05

    # Perform a two-sided Z-test
    results_two_sided = z_test_one_sample(sample, mu_0, sigma, alpha=alpha_level, alternative='two-sided')
    print("Two-sided Z-test results:")
    print(f"  Z-statistic: {results_two_sided['z_statistic']:.4f}")
    print(f"  P-value: {results_two_sided['p_value']:.4f}")
    print(f"  Decision: {results_two_sided['decision']}")

    print("\n" + "="*30 + "\n")

    # Perform a one-sided (greater) Z-test
    results_greater = z_test_one_sample(sample, mu_0, sigma, alpha=alpha_level, alternative='greater')
    print("One-sided (greater) Z-test results:")
    print(f"  Z-statistic: {results_greater['z_statistic']:.4f}")
    print(f"  P-value: {results_greater['p_value']:.4f}")
    print(f"  Decision: {results_greater['decision']}")
```

```python
# `16.  Create a confidence interval for a dataset using Python and interpret the result?

import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):

    sample_mean = np.mean(data)
    sample_std_dev = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
    sample_size = len(data)

    # Calculate the standard error of the mean
    standard_error = sample_std_dev / np.sqrt(sample_size)

    # Calculate the t-critical value for the given confidence level and degrees of freedom
    # Degrees of freedom (df) = sample_size - 1
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    # Calculate the margin of error
    margin_of_error = t_critical * standard_error

    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example Usage:
if __name__ == "__main__":
    dataset = [10, 12, 11, 13, 15, 14, 12, 16, 10, 13]

    lower_ci, upper_ci = calculate_confidence_interval(dataset)

    print(f"Dataset: {dataset}")
    print(f"Sample Mean: {np.mean(dataset):.2f}")
    print(f"95% Confidence Interval: ({lower_ci:.2f}, {upper_ci:.2f})")

    # Interpretation
    print("\nInterpretation:")
    print(f"Based on this sample, we are 95% confident that the true population mean")
    print(f"for the characteristic measured in the dataset lies between {lower_ci:.2f} and {upper_ci:.2f}.")
    print("This means that if we were to repeatedly take samples and construct confidence intervals,")
    print("approximately 95% of those intervals would contain the true population mean.")
```

```python

# 24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution?

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution(size=1000, mean=0, std_dev=1):

    # Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate skewness
    skewness = skew(data, bias=False)

    # Calculate kurtosis (Fisher's definition; normal distribution => 0.0)
    kurtosis_value = kurtosis(data, fisher=True, bias=False)

    return skewness, kurtosis_value

# Example usage:
skewness, kurtosis_value = analyze_normal_distribution()
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis_value}")
```
