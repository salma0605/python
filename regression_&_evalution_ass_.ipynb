1. What is Simple Linear Regression?

Ans:Simple linear regression is a statistical method used to model the relationship between a
dependent variable and a single independent variable using a straight line. It aims to find the
best-fitting line that represents the correlation between the two variables, allowing for
prediction and estimation.  

Key Concepts:

 - Dependent Variable:
The variable being predicted or whose values are being explained (often denoted as 'y').  

 -  Independent Variable:
The variable is used to predict the dependent variable (often denoted as 'x').

 - Linear Relationship:
The assumption that the relationship between the variables can be represented by a straight line.


 -  Regression Line:
The line of best fits through the data points that minimize the difference between predicted and
actual values.  

 - Slope:
The steepness of the regression line, indicating how much the dependent variable changes for each
unit change in the independent variable.  

 - Intercept:
The point where the regression line crosses the y-axis represents the value of the dependent variable
when the independent variable is zero.

How it Works:

1. Data Collection: Data is collected for both dependent and independent variables.

2. Model Fitting: The simple linear regression model is applied to the data to find the equation
of the best-fitting line (y = a + bx), where 'a' is the intercept and 'b' is the slope.  

3. Prediction: Once the model is established, it can be used to predict the value of the
dependent variable for a given value of the independent variable.  

Example:
Imagine trying to predict a student's final exam score (dependent variable) based on the number of
hours they studied (independent variable). Simple linear regression could be used to find the
relationship between these two variables, allowing you to estimate a student's potential score based
on their study hours.


Simple linear regression provides a way to quantify and understand the relationship between two
variables, allowing for predictions and insights into how changes in one variable impact the other.


2. What are the key assumptions of Simple Linear Regression?  


Ans:The key assumptions of simple linear regression are linearity, independence,
homoscedasticity, and normality. These assumptions ensure that the model accurately
reflects the relationship between the independent and dependent variables and that the
results are reliable.  

Here's a breakdown of each assumption:

1. Linearity:
A linear relationship exists between the independent variable (X) and the dependent variable
(Y). In other words, the relationship can be approximated by a straight line.  

2. Independence:
The residuals (the differences between observed and predicted values) are independent of
each other. This means that the error for one observation does not influence the error for any
other observation.  

3. Homoscedasticity:
The variance of the residuals is constant across all levels of the independent variable. In
simpler terms, the spread of the data points around the regression line should be roughly the
same for all values of X.


4. Normality:
For any given value of X, the dependent variable Y is normally distributed. This means that
the distribution of the errors is also normal. These assumptions are crucial for the validity of the linear regression model. If these
assumptions are violated, the model's predictions may be inaccurate, and the statistical tests
performed on the model may not be reliable.  

5. No (or Minimal) Measurement Error in X:
The independent variable (X) should be measured without significant error.
In practice, some error is acceptable, but high measurement errors in X can distort results.

6. No Significant Outliers or High Leverage Points
Outliers can skew the regression line.
High-leverage points (far from the mean of X) can have a large influence on the slope.

3. What is heteroscedasticity, and why is it important to address in regression
models?

Ans:Heteroscedasticity refers to the unequal variance of errors (residuals) across different levels
of an independent variable in a regression model. In simpler terms, it means the spread of the
data points around the regression line is not consistent. This violates a key assumption of
ordinary least squares (OLS) regression, which assumes homoscedasticity (constant variance
of errors). Addressing heteroscedasticity is crucial because it can lead to unreliable statistical
inferences and inaccurate predictions.  

Here's why it's important to address heteroscedasticity:

 -  Invalidated Hypothesis Tests:
Heteroscedasticity can lead to inaccurate standard errors, making hypothesis tests
unreliable. This means you might incorrectly reject or fail to reject a null hypothesis, leading to
flawed conclusions about the relationships between variables.  

 -  Inefficient Parameter Estimates:
While the parameter estimates (coefficients) in a regression model might still be unbiased in
the presence of heteroscedasticity, they are not the most efficient. This means there are other
estimators that could provide more precise estimates.  

 -  Misleading Confidence Intervals:Confidence intervals calculated with standard errors that are incorrect due to
heteroscedasticity will be inaccurate. This can lead to underestimation of the true range of
values for the parameters.  

• Poor Predictions:
When heteroscedasticity is present, predictions for certain ranges of the independent variable
(where the variance is higher) might be less accurate.  
Essentially, heteroscedasticity undermines the reliability and validity of regression models,
potentially leading to incorrect conclusions and poor decision-making. Therefore, it's important
to detect and address it using techniques like robust standard errors or generalized least
squares.  

How to Detect Heteroscedasticity

 -  Residuals vs Fitted Plot: Look for patterns like fans or cones.

 -  Breusch-Pagan Test: A formal statistical test for heteroscedasticity.

 - White Test: Another common test that doesn't assume a specific form of heteroscedasticity.

How to Fix It

1. Transform the Dependent Variable
o e.g., Use log(Y), square root(Y), or other transformations to stabilize variance.

2. Weighted Least Squares (WLS)
o Gives less weight to observations with higher variance.

3. Robust Standard Errors (Heteroscedasticity-Consistent SEs)
o Use tools like Huber-White standard errors to correct inference without changing the
model.

4. What is Multiple Linear Regression?  

Ans:Multiple linear regression is a statistical method used to model the relationship between one
dependent variable and two or more independent variables. It aims to find the best-fitting
linear equation that predicts the value of the dependent variable based on the values of the
independent variables. This technique is widely used in various fields for prediction, pattern
detection, and understanding the influence of multiple factors on a single outcome.  
Here's a more detailed explanation:

Key Concepts:

 - Dependent Variable:
The variable you are trying to predict (also known as the response variable).  

 - Independent Variables:
The variables used to predict the dependent variable (also known as predictors or explanatory
variables).

 - Linear Relationship:
The assumption that the relationship between the dependent and independent variables can
be represented by a straight line (or a plane in multiple dimensions).  

 - Regression Equation:
The equation that describes the linear relationship. For multiple linear regression, it takes the
form:  

Y = b₀ + b₁X₁ + b₂X₂ + ... + bₚXₚ + ε

Where:
 - Y is the dependent variable.

 - b₀ is the intercept (the value of Y when all independent variables are zero).  

 -  b₁, b₂, ... bₚ are the coefficients for each independent variable (X₁, X₂, ... Xₚ).

 -  X₁, X₂, ... Xₚ are the independent variables.  

 - ε is the error term (representing the unexplained variation).

 - Coefficients (b₁, b₂, ... bₚ): These values indicate the change in the dependent variable for a
one-unit change in the corresponding independent variable, holding all other independent

variables constant.

 -  Model Fit:
It's important to assess how well the model fits the data using metrics like R-squared and p-values.

 - Error Term (ε): Represents the random variation or noise in the data that is not explained by the model.


How it works:
Multiple linear regression uses a process called "least squares" to find the best-fitting line (or plane in
multiple dimensions). It minimizes the sum of the squared differences between the observed values of
the dependent variable and the values predicted by the regression equation.

Applications:

Multiple linear regression is used in:

 - Predicting sales based on advertising spend, price, and competitor activity.

 - Forecasting stock prices based on various economic indicators.

 -  Modeling disease progression is based on multiple risk factors.

 -  Understanding the impact of different factors on student performance.

Important Considerations:

  - Assumptions:
Multiple linear regression relies on certain assumptions about the data, such as linearity,
independence of errors, and constant variance of errors.

- Multicollinearity:
High correlation between independent variables can make it difficult to interpret the individual effects
of each variable.  

- Model Fit:
It's important to assess how well the model fits the data using metrics like R-squared and p-values.

5. What is polynomial regression, and how does it differ from linear
regression?


Ans:Polynomial regression is a type of regression analysis that models the relationship between a
dependent variable and an independent variable as an nth degree polynomial. Unlike linear
regression, which uses a straight line to model the relationship, polynomial regression can
model non-linear relationships by fitting a curve to the data.  

Key Differences:

 - Linear Regression:
Models a linear relationship between variables using a straight line. The equation is typically
in the form of y = mx + b, where 'm' is the slope and 'b' is the y-intercept.

 -  Polynomial Regression:
Models a non-linear relationship using a polynomial equation of a specified degree. This
allows for fitting curves to the data, capturing more complex relationships.

 - Flexibility:
Polynomial regression is more flexible than linear regression because it can fit a wider range
of data patterns.

 -  Complexity:
Higher-degree polynomial models can become more complex and prone to overfitting,
meaning they might fit the training data very well but generalize poorly to new data.

 - Feature Engineering:
Polynomial regression effectively creates new features (e.g., squaring or cubing the original
input) to capture non-linear relationships, while linear regression treats each feature as
independent.


If your data points follow a straight line, linear regression is sufficient. However, if your data
points form a curve, polynomial regression, with its ability to fit curves, is a better choice.  

When to Use Polynomial Regression:

 -  When the relationship between variables is not linear but still deterministic.
  
 -  Useful for curved trends in data, like population growth, disease spread, sales cycles,
etc.

```python
# 6. :  Implement a Python program to fit a Simple Linear Regression model to the following sample data:
# ●	X = [1, 2, 3, 4, 5]
# ●	Y = [2.1, 4.3, 6.1, 7.9, 10.2]

# Plot the regression line over the data points.


import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values using the model
Y_pred = model.predict(X)

# Print model parameters
print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_[0]}")

# Plotting
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 7. : Fit a Multiple Linear Regression model on this sample data:
#●	Area = [1200, 1500, 1800, 2000]
#●	Rooms = [2, 3, 3, 4]
#●	Price = [250000, 300000, 320000, 370000]
# Check for multicollinearity using VIF and report the results.

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Sample Data
data = pd.DataFrame({
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
})

# Define independent and dependent variables
X = data[['Area', 'Rooms']]
y = data['Price']

# Fit the Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Print model coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Add constant term for VIF calculation
X_with_const = sm.add_constant(X)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i)
                   for i in range(X_with_const.shape[1])]

print("\nVariance Inflation Factors (VIF):")
print(vif_data)


```

```python
# 8 Implement polynomial regression on the following data: ● X = [1, 2, 3, 4, 5]
# ●	Y = [2.2, 4.8, 7.5, 11.2, 14.7]
# Fit a 2nd-degree polynomial and plot the resulting curve.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Transform X to include polynomial terms (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict Y values
X_range = np.linspace(1, 5, 100).reshape(-1, 1)  # Smooth range for curve
X_range_poly = poly.transform(X_range)
Y_pred = model.predict(X_range_poly)

# Print model coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")

# Plot the results
plt.scatter(X, Y, color='blue', label='Original Data')
plt.plot(X_range, Y_pred, color='red', label='Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2nd-Degree Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# 9. 9: Create a residuals plot for a regression model trained on this data:
# ●	X = [10, 20, 30, 40, 50]
# ●	Y = [15, 35, 40, 50, 65]
# Assess heteroscedasticity by examining the spread of residuals.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Train the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Print coefficients and residuals
print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_[0]}")
print(f"Residuals: {residuals}")

# Plot residuals
plt.scatter(X, residuals, color='purple')
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.grid(True)
plt.show()

```
